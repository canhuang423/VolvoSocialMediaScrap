```python
review_title=[]
car_model_year=[]
review=[]
car_model = [
    '240',
    '740',
    '760',
    '780',
    '850',
    '940',
    '960',
    'C30',
    'C70',
    'Coupe',
    'S40',
    'S70',
    'S80',
    'V40',
    'V50',
    'V70',
    'XC',
    'XC70',
    'S60',
    'S90',
    'V60',
    'V90',
    'XC40',
    'XC60',
    'XC90',
]

car_year=['2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']
```

```python
        # run chrome webdriver from executable path of your choice
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.maximize_window()
```

```python
import requests

for model in car_model:
    for year in car_year:
        urlpage = 'https://www.kbb.com/volvo/' + model + '/' + year + '/' + 'consumer-reviews/'
        print(model+" - " + year + " - ")
        # print(urlpage)
        
        response = requests.get(urlpage,headers={"User-Agent": "Mozilla/5.0 (X11; CrOS x86_64 12871.102.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.141 Safari/537.36"})
        html_soup = BeautifulSoup(response.text, "html.parser")
        text = html_soup.find('div', attrs={'id': 'root'})

        if (text.text != ''):
            driver.get(urlpage)
            # driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            # elem = driver.find_element_by_css_selector("button.css-1ad45u5-commonStyle-default-secondary")
            # driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            # elem.click()
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            
            # review title
            titles= driver.find_elements_by_xpath("//div[@class='css-4eaujp']")
            # loop over results
            for title in titles:
                product_name = title.text
                review_title.append(product_name)
            review_title.pop(0)
            
            #take the date and car model
            date_models = driver.find_elements_by_xpath("//div[@class='css-1wnwnp8-LabelText euqiqzc1']")
            # loop over results
            for date_model in date_models:
                product_name = date_model.text
                car_model_year.append(product_name)

            #take the review 
            reviews=driver.find_elements_by_xpath("//div[@class='css-3u9dpw-ColBase eib3s4b0']")
            # loop over results
            for re in reviews:
                #rev=re.find_elements_by_xpath("//p[@class='css-1jqpogp-StyledParagraph e12faeex0']")
                product_name = re.text
                review.append(product_name)

```
