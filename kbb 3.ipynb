```python
# import libraries
import pandas as pd
import numpy as np
import urllib.request
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
# specify the url
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException

```

```python
review_title=[]
car_model_year=[]
review=[]
car_model = [
    '240',
    '740',
    '760',
    '780',
    '850',
    '940',
    '960',
    'C30',
    'C70',
    'Coupe',
    'S40',
    'S70',
    'S80',
    'V40',
    'V50',
    'V70',
    'XC',
    'XC70',
    'S60',
    'S90',
    'V60',
    'V90',
    'XC40',
    'XC60',
    'XC90',
]

car_year=['2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']
```

```python
for model in car_model:
    for year in car_year:
        urlpage = 'https://www.kbb.com/volvo/' + model + '/' + year + '/' + 'consumer-reviews/'
        print(model+" - " + year + " - ")
        print(urlpage)
        
        # run chrome webdriver from executable path of your choice
        driver = webdriver.Chrome(ChromeDriverManager().install())
        driver.maximize_window()
        
        
        try:
            driver.get(urlpage)
        # find the element's css name
            elem = driver.find_element_by_css_selector("button.css-1lsmfli-commonStyle-default-secondary")
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            # ask javascript to clikc on the button
            elem.click()
            # ask javascript to scroll down to new bottom to get full page
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            
        except NoSuchElementException:
            print('Element BY3EC not found') # or do something else here
            
            
   
        
        # review title
        titles= driver.find_elements_by_xpath("//div[@class='css-4eaujp']")
        # loop over results
        for title in titles:
            product_name = title.text
            review_title.append(product_name)
        review_title.pop(0)
        
        #take the date and car model
        date_models = driver.find_elements_by_xpath("//div[@class='css-1wnwnp8-LabelText euqiqzc1']")
        # loop over results
        for date_model in date_models:
            product_name = date_model.text
            car_model_year.append(product_name)

        #take the review 
        reviews=driver.find_elements_by_xpath("//div[@class='css-3u9dpw-ColBase eib3s4b0']")
        # loop over results
        for re in reviews:
            #rev=re.find_elements_by_xpath("//p[@class='css-1jqpogp-StyledParagraph e12faeex0']")
            product_name = re.text
            review.append(product_name)
            
```
